```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Space to plot things
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)

# Photon Source
ax.add_patch(patches.Circle((1, 3), 0.3, color="purple"))
ax.text(1, 2.4, "Photon\nSource", ha="center", color="black")

# Beam Splitter
beam_splitter = patches.Rectangle((3.5, 2.5), 0.3, 1, angle=45, color="gray")
ax.add_patch(beam_splitter)
ax.text(4, 3.5, "Beam Splitter", ha="center", color="black")

# Detectors
ax.add_patch(patches.Circle((7, 5), 0.3, color="blue"))
ax.text(7, 5.5, "Detector 1", ha="center", color="black")
ax.add_patch(patches.Circle((7, 1), 0.3, color="blue"))
ax.text(7, 1.5, "Detector 2", ha="center", color="black")

# Draw light from photon and then to Detectors
plt.plot([1.3, 3.7], [3, 3], "k--", lw=1)
plt.plot([3.8, 7], [3.8, 5], "k-", lw=1)
plt.plot([3.8, 7], [2.2, 1], "k-", lw=1)

# Arrow indicating photon flow
arrow1 = patches.FancyArrowPatch((1.5, 3), (3.4, 3), mutation_scale=10, color="red", linewidth=1)
arrow2 = patches.FancyArrowPatch((3.8, 3.7), (6.7, 4.8), mutation_scale=10, color="red", linewidth=1)
arrow3 = patches.FancyArrowPatch((3.8, 2.3), (6.7, 1.2), mutation_scale=10, color="red", linewidth=1)

ax.add_patch(arrow1)
ax.add_patch(arrow2)
ax.add_patch(arrow3)

# Remove ticks
ax.set_xticks([])
ax.set_yticks([])

# Set aspect
ax.set_aspect('equal')
plt.title("Schematic of the Hanbury Brown and Twiss Experiment")
plt.show()
```
